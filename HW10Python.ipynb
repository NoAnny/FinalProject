```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data['I_am_human']=0
data['I_am_robot']=0
data.head()
```

```python
data.loc[(data['whoAmI'] == 'human'), 'I_am_human'] = 1
data.loc[(data['whoAmI'] == 'robot'), 'I_am_human'] = 0
data.loc[(data['whoAmI'] == 'human'), 'I_am_robot'] = 0
data.loc[(data['whoAmI'] == 'robot'), 'I_am_robot'] = 1
data
```
